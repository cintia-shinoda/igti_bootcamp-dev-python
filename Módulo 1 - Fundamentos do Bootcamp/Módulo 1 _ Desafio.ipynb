# Q9

Qual o propósito do ```break``` no bloco de código 6?

````python
# Bloco de Código 6

n = 5
while n > 0:
    n -= 1
    if n == 2:
        break
    print(n)
print('Fim do Loop')
````

````python
# Resposta: Parar o laço de repetição quando um determinado valor é alcançado.
````

# Q10

Qual o propósito do ```continue``` no bloco de código 7?

````python
# Bloco de Código 7

n = 5
while n > 0:
    n -= 1
    if n == 2:
        continue
    print(n)
print('Fim do Loop')
````

````python
# Resposta: Continuar o laço de repetição quando um valor é alcançado. Ignorando esse mesmo valor.
````
