````python
"""
Exemplo:
Solução de um sistema de equações:
    a + 2*b = 7
  3*a - 2*b = -11

  solução analítica (a,b) = (-1, 4)

Matricialmente, este problema tem a seguinte forma:
    Ax = C, onde:
    - x = [a,b]
    - A = [[1,2], [3, -2]]
    - C = [[7, -11]]
    solução numérica: x = inv(A) @ C,
"""

# definição do problema
A = np.array([[1, 2], [3, -2]])
C = np.array([[7], [-11]])

print("A:\n", A)
print("C:\n", C)
````

````python
# solução:
# x = np.dot(np.linalg.inv(A), C)
x = np.linalg.inv(A) @ C

print("(a, b):", x.ravel())
````

## 2.8 - Comparações e Indexação Booleana

### Comparações Booleanas

operações elemento a elemento

````python
A = np.array([1, 2, 3])
B = np.array([2, 0, 2])
s = 3

# menor
print("Comparação menor:")
print(A < B)
print(A < s)  # broadcasting
print()

# menor ou igual
print("Comparação menor ou igual:")
print(A <= B)
print(A <= s)  # broadcasting
print()

# maior
print("Comparação maior:")
print(A > B)
print(A > s)  # broadcasting
print()

# maior ou igual
print("Comparação maior ou igual:")
print(A >= B)
print(A >= s)   # broadcasting
print()

# igual
print("Comparação de igualdade:")
print(A == B)
print(A == s)    # broadcasting
````

## 2.11 - Regressão Linear no numpy: conceitos básicos

**Problema:** dado um conjunto de pontos, queremos achar qual a função que melhor descreve esses pontos

Diversas possíveis soluções... Qual função escolher?

- a escolha da melhor função deve ser baseada em um critério
- um critério comumente utilizada é o *erro quadrático*, que queremos minimizar

(quadrático: não importa se erro é para mais ou para menos)

- com o critério definido e, sabendo que a função linear possui o seguinte formato:

y = f(x) = ax + b

- o problema de regressão resume-se à determinação dos coeficientes **a** e **b**, visto que  **x** e **y** são dados de entrada

## 2.12 - Regressão Linear no numpy - Prática

````python
# visualização de dados
import matplotlib.pyplot as plt
````

````python
# dados
x = [-1., -0.77777778, -0.55555556, -0.33333333, -0.11111111, 0.11111111, 0.33333333, 0.55555556, 0.77777778, 1.]
y = [-1.13956201, -0.57177999, -0.21697033, 0.5425699, 0.49406657, 1.14972239, 1.64228553, 2.1749824, 2.64773614, 2.95684202]
````

````python
# plot dos dados
plt.figure(figsize=(10,5))
plt.plot(x, y, 'o', label = 'dados originais')  # scaterplot
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
````

pseudo-inversa de X, calculada pelo método ```np.linalg.pinv(x)```

````python
"""
Iremos estimar uma função do tipo: y = a*x + b
ou seja, devemos achar quais os valores de a e b
que melhor representam os dados

Os valores reais de a e b são: (a, b): 2, 1
"""

# transformando para numpy o vetor coluna
x, y = np.array(x).reshape(-1, 1), np.array(y).reshape(-1,1)

# adicionando bias: para estimar o termo b
X = np.hstack((x, np.ones(x.shape)))

# estimando a e b
beta = np.linalg.pinv(X).dot(y)
print("a estimado:", beta[0][0])
print("b estimado:", beta[1][0])
````

````python
# plot dos dados
plt.figure(figsize=(10,5))
plt.plot(x, y, 'o', label = 'dados originais')
plt.plot(x, X.dot(beta), label='regressão linear')
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Regressão linear com numpy")
plt.grid()
plt.show()
````

````python
df
````

###```inplace```

````python
# cópia de um dataframe: evita compartilhamento de memória
# sem copy(), operaçõees inplace em df2 também alteram df
df2 = df.copy()
````

````python
# argumento inplace
# inplace = True aplica a transformação no próprio objeto
df2.drop('temperatura', axis=1)
````

````python
# sem inplace, df2 continua o mesmo
df2
````

````python
df3 = df2.drop("temperatura", axis=1)
df3
````

````python
# para aplicar a modificação no próprio df2:
# inplace=True aplica a transformação no próprio objeto
df2.drop('temperatura', axis=1, inplace=True)
df2
````

````python
# o objeto original segue inalterado, pois fez-se uma cópia
df
````
