Qual estação do ano contém a maior média de locações de bicicletas?

## L
Que dia da semana contém a maior média de locações de bicicletas?

````python
mean_weekday = df.groupby('weekday').mean('total_count')
print(mean_weekday)
# 0: domingo, 1: segunda-feira, …, 6: sábado

# resposta: 4: quinta-feira
````

## N
Às quartas-feiras (```weekday = 3```), qual o horário do dia contém a maior média de locações de bicicletas?

````python
df_wednesday = df.loc[df.weekday==3]
print(df_wednesday)
````

````python
df_wednesday_by_hour = df_wednesday.groupby('hour').mean('total_count')
print(df_wednesday_by_hour)

# resposta: 17h
````

## O

````python
df_saturday = df.loc[df.weekday==6]
df_saturday.head(10)
````

````python
df_saturday_by_hour = df_saturday.groupby('hour').mean('total_count')
print(df_saturday_by_hour)

# resposta: 13h
````
