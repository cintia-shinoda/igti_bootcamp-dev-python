Nesta aula:

### Protocolo HTTP

- Requisição HTTP
    - Método HTTP
    - Cabeçalho: servidor, formato, parâmetros
    - Corpo

- Processamento da Requisição

- Resposta HTTP
    - Status HTTP
    - Cabeçalho
    - Conteúdo


**Tipos de requisições HTTP:** 
(método)
- **GET** 
    - tipo mais comum de solicitação na web
    - simplesmente recuperam dados de um servidor
    - 99% das vezes, são usados para solicitar HTML
- **POST**
    - usada para enviar dados ao servidor e criar um novo objeto (entidade) no servidor
    - os dados são enviados no *body* da requisição
- **PUT**
    - usadas para enviar dados ao servidor para substituir um objeto (entidade) no servidor
    - exemplo: substituir video
- **PATCH**
    - usada para atualizar algum atributo do objeto/entidade existente no servidor
    - exemplo: descrição do vídeo
- **DELETE**
    - semelhante à requisição GET por não terem *body*
    - é o oposto de GET
    - usada para excluir dados do servidor

**Códigos de Status**
- 1xx : Informação
- 2xx : Sucesso
- 3xx : Redirecionamento
- 4xx : Erro no Cliente
- 5xx : Erro no Servidor

### O que é um formulário HTML?

- principal ponto de interação entre um usuário e um website ou aplicativo
- permite que os usuários enviem dados para um servidor web, mas a página web também pode usá-los por conta própria
- é feito de um ou mais widgets:
    - campos de texto
    - caixas de seleção
    - botões
    - checkboxes
    - radio buttons

### O elemento ```<form>```

**action**
este atributo define o local (url) em que os dados do formulários devem ser enviados

**method**
este atributo define qual o método HTTP para enviar os dados ("GET" ou "POST")
<br><br>

Exemplo:
```html
<form action="servidor-que-processa-dados-do-form" method="post">Meu Formulário</form>
```

### Enviar os dados para um Servidor Web



- para lidar com dados de formulário no lado do servidor, precisamos dar nome aos nossos dados (atributo "name")
- os nomes permitem que o servidor lide com cada campo pelo nome

```html
<form action="/servidor" method="post">
    <div>
        <label for="nome">Nome:</label>
        <input type="text" id="nome" name="usuario_nome"/>
    </div>

    <div>
        <label for="email">E-mail:</label>
        <input type="email"id="email" value="seu@email.com" name="usuario_email"/>
    </div>

    <div>
        <label for="msg">Mensagem:</label>
        <textarea id="msg" name="usuario_msg">Área de Texto:</textarea>
    </div>

    <div>
        <button type="submit">Enviar Mensagem</button>
        <button type="reset">Resetar Campos</button>
    </div
</form>
```

- o formulário irá enviar 3 informações (usuario_nome, usuario_email e usuario_msg) para a url /servidor com o método POST
- no lado do servidor, existe um código para tratar a url /servidor, que receberá os dados como uma lista de 3 itens chave/valor contidos na solicitação HTTP

### Introdução a Web Services

- Serviço HTTP/Web
- solução utilizada na integração de sistemas
- aplicações que fornecem serviços simples e que poodem interagir entre si
- se comunicam em uma rede por meio de HTTP
- todos os serviços web são APIs, mas nem todas as APIs são serviços da web

### Fluxo SOAP
- Simple Object Access Protocol
<br><br><br>

- baseado no XML
- Adotado pelo W3C
- Estrutura extensível de mensagens
- Meensagem via HTTP ou SMTP
<br><br><br>

-  WSDL: Web Services Description Language


### Fluxo REST

- Representational State Transfer
<br><br><br>

- Geralmente JSON
- Arquitetura de software (conjunto coordenado de restrições)
- Transações aproveitam o protocolo HTTP (GET, POST, DELETE,...)
- Stateless - o servidor não armazena informações
<br><br><br>

- API Gateway

### Comparação: REST x SOAP


REST (Representational State Transfer) | SOAP (Simple Object Access Protocol)
---  | --- |
Estilo de arquitetura de software | Protocolo de comunicação |
Abordagem mais simples | Maior complexidade |
Menos burocrática | Especificação bem definidas |
Fortemente relacionado ao HTTP | HTTP como meio de transporte |
Muitos recursos (URIs) | Poucas (URIs) |
Métodos uniformes | Muitos métodos customizados |
Suporta JSON, XML, YAML, ... | Suporta XML |
Pode ler cache |  Invoca serviços chamando o método RPC |
Fácil integração com JavaScript
Amplamente e frequentemente usado | Desempenho não é muito grande comparado ao REST|
<br><br><br>

Quando usar qual:

REST | SOAP |
--- | --- |
Quando a sua largura de banda é muito limitada | Ao executar operações que envolvam várias chamadas |
Quando o cliente e o servidor operam em um ambiente web | Quando você quiser fazer cumprir um contrato estrita entre cliente e servidor |
Exemplos: Serviço de Social Media, Serviço de Chat na Web | Exemplos: Serviços Financeiros, Serviços de Telecomunicações |



## 3.2 - Introdução ao Python WSGI (Web Server Gateway Interface)

Nesta aula:
- Aplicações Web em Python
- Objetivos do WSGI

### Aplicações Web em Python

- um servidor web tradicional não entende código Python
- uma aplicação web em Python *em produção* possui:
    - **Servidor Web:** hospeda *arquivos estáticos* (HTML, CSS, JS) e recebe conexões HTTP (exemplos: Apache, NGINX)
    - **Servidor WSGI:** *invoca processos/instâncias* de aplicação web e distribui carga (exemplo: gunicorn)
    - **Aplicação Web:** aplicação usando um *framework web* (páginas dinâmicas, APIs)

### Objetivos do WSGI

- Web Server Gateway Interface ~ "porta de entrada do servidor web"
- Por que usar  WSGI e não apenas apontar um servidor web para um app?
    - **Flexibilidade (PEP 3333):** um desenvolvedor pode mudar de Unicorn para uWSGI sem modificar uma aplicação que implementa WSGI
    - **Dimensionamento:** atender milhares de requisições de conteúdo dinâmico de uma vez é a responsabilidade dos servidores WSGI
